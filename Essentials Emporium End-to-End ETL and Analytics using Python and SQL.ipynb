```python
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df.head(20)
```

```python
df['Ship Mode'].unique()
```

```python
#df.rename(columns={'Order Id':'order_id', 'City':'city'})
df.columns=df.columns.str.lower() #lowering the case of column names
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
df['discount']=df['list_price']*df['discount_percent']*.01
df.head(5)
```

```python
df['sale_price']= df['list_price']-df['discount']
df.head(5)
```

```python
df['profit']=df['sale_price']-df['cost_price']
df.head(5)
```

```python
df.dtypes
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d") #converting data type of column "Date" from object to datetime
df.head(5)
```

```python
df.dtypes
```

```python
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df.head(20)
```

```python
import sqlalchemy as sal
engine = sal.create_engine('mssql+pyodbc://localhost/master?driver=ODBC+Driver+17+for+SQL+Server')
conn = engine.connect()
```

```python
df.to_sql('df_orders', con=conn , index=False, if_exists = 'append') #loading the data into sql server using append option
```
